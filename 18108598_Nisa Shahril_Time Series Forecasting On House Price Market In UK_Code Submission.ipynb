## Final Data

```python
#functools module is for functions that work on other functions
#reduce function accepts a function and sequence and returns a single value calculated
#reduce is moved to functools module, hence the code

from functools import reduce
```

```python
#this is to reset the index of each dataframe
#drop: do not try to insert index into dataframe columns, this resets the index to the default integer index
#inplace: modify the dataFrame in place (do not create a new object)

concatenate_AEI.reset_index(drop=True, inplace=True)
gdp_new.reset_index(drop=True, inplace=True)
durable_goods_new.reset_index(drop=True, inplace=True)
house_eq_wid_new.reset_index(drop=True, inplace=True)
household_savings_ratio_new.reset_index(drop=True, inplace=True)
bankrate_new.reset_index(drop=True, inplace= True)
claimant_count_new.reset_index(drop=True, inplace= True)
retail_price_index_new.reset_index(drop=True, inplace=True)
houseprice_index_new.reset_index(drop = True, inplace = True)
```

```python
#delete unused columns for the dataset

durablegoods_new=durable_goods_new.drop(columns={'Quarterly'})
householdSavingsRatio_new=household_savings_ratio_new.drop(columns={'Quarters'})
```

```python
#rename columns for the merge dataset

house_eq_wid_new=house_eq_wid_new.rename(columns={"Changes": "House Equity Withdrawal Changes"})
gdp_new=gdp_new.rename(columns={"%": "GDP Percent"})
concatenate_AEI=concatenate_AEI.rename(columns={"Index": "AEI Index"})
bankrate_new=bankrate_new.rename(columns={"Rate": "Bank Rate Value"})
durablegoods_new=durablegoods_new.rename(columns={"Total": "Durable Goods Total"})
householdSavingsRatio_new=householdSavingsRatio_new.rename(columns={"Percent": "Household Savings Ratio Percent"})
claimant_count_new=claimant_count_new.rename(columns={"Rate": "Claimant Count Rate"})
retail_price_index_new= retail_price_index_new.rename(columns={"%": "RPI Percent"})
houseprice_index_new= houseprice_index_new.rename(columns={"Index": "House Price Index"})
```

```python
concatenate_AEI.head()
```

```python
durablegoods_new.head()
```

```python
gdp_new.head()
```

```python
house_eq_wid_new.head()
```

```python
householdSavingsRatio_new.head()
```

```python
bankrate_new.head()
```

```python
claimant_count_new.head()
```

```python
houseprice_index_new.head()
```

```python
#This is to merge the data sets. 
#on: is column or index level anmes to join on
#how, outer: type of merge to be performed; use union of keys from data, sort keys lexicographically (dictionary order)


merge_data = reduce(lambda x,y: pd.merge(x,y, on= 'Date', how= 'outer'), 
                    [gdp_new, concatenate_AEI, house_eq_wid_new, durablegoods_new, householdSavingsRatio_new, bankrate_new, 
                     claimant_count_new, retail_price_index_new, houseprice_index_new])
```

```python
merge_data = merge_data.set_index('Date')
```

```python
merge_data.info()
```

```python
#Fill NA/NaN values using method
#bfill: use next valid observation to fill gap

merge_data = merge_data.fillna(method ='bfill')
```

```python
#eleminate rows where there are any missing elements in the data
merge_data = merge_data.dropna()
```

```python
file_name = 'MergeData1.csv'

# saving the excel
merge_data.to_csv(file_name)
```

# Multivariate Time Series with RNN model using LSTM

```python
#Perform imports and load datasets

import pandas as pd
import numpy as np
%matplotlib inline
%config InlineBackend.figure_format ='retina'
import matplotlib.pyplot as plt
import statsmodels.api as sm
from pandas import DataFrame, concat
from sklearn.metrics import mean_absolute_error, mean_squared_error
from numpy import mean, concatenate
from math import sqrt
from sklearn.preprocessing import MinMaxScaler, StandardScaler
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense,LSTM, Activation, Dropout
from numpy import array, hstack
from tensorflow import keras
import tensorflow as tf

#rnndf = pd.read_csv("MergeData1.csv", index_col=['Date'], infer_datetime_format = True)

rnndf = pd.read_csv("MergeData1.csv")
rnndf = rnndf.drop(['GDP Percent', 'House Equity Withdrawal Changes', 'Household Savings Ratio Percent', 
                                  'Claimant Count Rate', 'RPI Percent'], axis = 1)
train_dates = pd.to_datetime(rnndf['Date'])

cols = list(rnndf)[1:6]
rnndf_training = rnndf[cols].astype(float)
rnndf_training
```

```python
rnndf.info()
```

```python
scaler = StandardScaler()
scaler = scaler.fit(rnndf_training)
rnndf_training_scaled = scaler.transform(rnndf_training)
```

```python
trainX=[] #training
trainY=[] #prediction

n_future = 1 #number of days we want to predict into the future
n_past = 12 #number of past days we want to use to predict the future

for i in range(n_past, len(rnndf_training_scaled) - n_future + 1):
    trainX.append(rnndf_training_scaled[i - n_past:i, 0:rnndf_training.shape[1]])
    trainY.append(rnndf_training_scaled[i + n_future -1:i + n_future, 0])
    
#converting list into array
trainX, trainY = np.array(trainX), np.array(trainY)

#trainX1 = tf.convert_to_tensor(trainX, dtype = tf.float32)
#trainY1 = tf.convert_to_tensor(trainY, dtype = tf.float32)

print('trainX shape == {}'.format(trainX.shape))
print('trainY shape == {}'.format(trainY.shape))
```

```python
#pip install -U numpy==1.18.5
```

```python
model = Sequential()
model.add(LSTM(100, activation='relu', input_shape=(trainX.shape[1], trainX.shape[2]), return_sequences = True))
model.add(LSTM(32, activation = 'relu', return_sequences = False))
model.add(Dropout(0.2))
model.add(Dense(trainY.shape[1]))

model.compile(optimizer = 'adam', loss = 'mse')
model.summary()
```

```python
#fit model
history = model.fit(trainX, trainY, epochs=10, batch_size = 16, validation_split = 0.1, verbose =1)
```

```python
#forecasting

n_future = 24
forecast_period_dates = pd.date_range(list(train_dates)[-1], periods= n_future, freq = '1m')

forecast = model.predict(trainX[-n_future:])
forecast_copies = np.repeat(forecast, rnndf_training.shape[1], axis = -1)

y_pred_future_AEI = scaler.inverse_transform(forecast_copies)[:,0]
y_pred_future_DG = scaler.inverse_transform(forecast_copies)[:,1]
y_pred_future_BR = scaler.inverse_transform(forecast_copies)[:,2]
y_pred_future_HPI = scaler.inverse_transform(forecast_copies)[:,3]
```
